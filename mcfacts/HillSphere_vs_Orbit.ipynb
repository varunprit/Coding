```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, ax = plt.subplots()
a = np.random.uniform(low=10.01, high=100, size=8)
bh = 1.e3
smbh = 1.e8
m = ((bh)/(smbh))**(1/3)
y = a*m
plt.scatter(x=a, y=y)
plt.xlabel('Black Hole Orbit')
plt.ylabel('R-Hill')
plt.title("Black Hole Orbit vs Hill Radius")
```
