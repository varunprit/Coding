```python
import matplotlib.pyplot as plt
```

```python
x= np.linspace(0,2*np.pi,100)
y=np.sin(x)
plt.plot(x,y,color='orange',linestyle='dashed')
```

```python
y2=np.cos(x)
y3=np.tan(x)
plt.plot(x,y2,label='cos')
plt.plot(x,y3,label='tan')
plt.legend()
plt.xlabel('degree')
plt.title('Trig Functions')
plt.ylim([-2,2])
```

```python
x=np.random.random(30)
y=0.5*x+0.1*np.random.random(30)
plt.scatter(x,y)
```

```python
import scipy.optimize as opt
def line(x,m,b):
    return m*x+b
xvals=np.linspace(0,2*np.pi,100)
popt,_=opt.curve_fit(line,x,y)
plt.plot(xval,line(xval,popt[0],popt[1]))
plt.scatter(x,y,marker='+',x='green',label='y')
plt.ylim([0,0.6])
```
